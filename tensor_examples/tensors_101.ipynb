## Broadcasting

```python
# PyTorch will allow us to multiply things that are the same shape, as well as shapes where one operand is of size 1 in a given dimension.
# It also appends the leading dimensions of size 1 automatically.
```

```python
torch.tensor([[1,2,3]]) + 10
```

```python
torch.tensor([[1, 2, 3]]) * torch.tensor([[10],[20],[30]]) # shape(1,3) * shape(3,1) = shape(3,3)
```

## Named tensors

```python
img_t = torch.randn(3, 5, 5) # shape(3,5,5) => shape [channels (RGB), rows, columns]
weights = torch.tensor([0.2126, 0.7152, 0.0722])
weights.shape
```

```python
batch_t = torch.randn(2, 3, 5, 5) # shape(2, 3, 5, 5) => shape [batch, channels (RGB), rows, columns]
```

```python
img_gray_naive = img_t.mean(-3) # lazy, unweighted mean on dimension -3 (channel)
img_gray_naive
```

```python
batch_gray_naive = batch_t.mean(-3) # again lazy, unweighted mean on dimension -3 (channel)
batch_gray_naive
```

```python
img_gray_naive.shape, batch_gray_naive.shape
```

```python
unsqueezed_weights = weights.unsqueeze(-1).unsqueeze(-1) # shape(3,1,1)
unsqueezed_weights
```

```python
img_weights = img_t * unsqueezed_weights # shape(3,5,5)
img_weights
```

```python
batch_weights = batch_t * unsqueezed_weights # shape(2,3,5,5) remember that broadcasting appends the leading dimensions of size 1 automatically
batch_weights
```

```python
img_gray_weighted = img_weights.sum(-3) # sum on third dimension from the end (channel), shape(5,5)
img_gray_weighted
```

```python
batch_gray_weighted = batch_weights.sum(-3) # sum on third dimension from the end (channel), shape(2,5,5)
batch_gray_weighted
```

```python
batch_weights.shape, batch_t.shape, unsqueezed_weights.shape
```

```python
weights_named = torch.tensor([0.2126, 0.7152, 0.0722], names = ['channels'])
weights_named
```

```python
print(weights_named.shape, weights_named.names)
```

```python
img_t_named = img_t.refine_names(..., 'channels', 'rows', 'columns')
batch_t_named = batch_t.refine_names(..., 'channels', 'rows', 'columns')
print("img named: ", img_t_named.shape, img_t_named.names)
print("batch named: ", batch_t_named.shape, batch_t_named.names)
```

```python
# For operations with two inputs, in addition to the usual dimension checks—whether sizes are the same, 
# or if one is 1 and can be broadcast to the other—PyTorch will now check the names for us. 
# So far, it does not automatically align dimensions, so we need to do this explicitly.
# The method align_as returns a tensor with missing dimensions added and existing ones permuted to the right order
```

```python
weights_named_aligned = weights_named.align_as(img_t_named)
weights_named_aligned.shape, weights_named_aligned.names
```

```python
# Functions accepting dimension arguments, like sum, also take named dimensions
gray_named = (img_t_named * weights_named_aligned).sum('channels')
gray_named
```

```python
img_t_named[...,:3]
```

```python
# gray_named = (img_t_named[...,:3] * weights_named).sum('channels') # doesn't work
```

```python
# If we want to use tensors outside functions that operate on named tensors, we need to drop the names by renaming them to None.
# The following gets us back into the world of unnamed dimensions
gray_plain = gray_named.rename(None)
gray_plain.shape, gray_plain.names
```

## Tensor element types. dtype

```python
# the objects within a tensor must all be numbers of the same type, and PyTorch must keep track of this numeric type.
```

```python
# torch.float32 or torch.float—32-bit floating-point
# torch.float64 or torch.double—64-bit, double-precision floating-point
# torch.complex64 or torch.cfloat—64-bit complex
# torch.complex128 or torch.cdouble—128-bit complex
# torch.float16 or torch.half—16-bit, half-precision floating-point
# torch.bfloat16—brain float 16-bit, half-precision floating-point (This is a new data type introduced by Google in 2019. It is similar to torch.float16 but uses a different encoding that can encode a larger range, while float16 is better for precision.)
# torch.int8—signed 8-bit integers
# torch.uint8—unsigned 8-bit integers
# torch.int16 or torch.short—signed 16-bit integers
# torch.int32 or torch.int—signed 32-bit integers
# torch.int64 or torch.long—signed 64-bit integers
# torch.bool—Boolean
# The default data type for tensors is 32-bit floating-point.
```

```python
double_points = torch.ones(10, 2, dtype=torch.double)
double_points
```

```python
short_points = torch.tensor([[1,2],[3,4]], dtype=torch.short)
short_points
```

```python
# casting
double_points = torch.zeros(10,2).double()
short_points = torch.ones(10,2).short()
double_points, short_points
```

```python
# or
double_points = torch.zeros(10,2).to(torch.double)
short_points = torch.ones(10,2).to(dtype=torch.short)
double_points, short_points
```

```python
# when mixing input types in operations, the inputs are converted to the larger type automatically
```

```python
points_64 = torch.rand(5, dtype=torch.double) # rand initializes the tensor elements to random numbers between 0 and 1.
points_short = points_64.to(torch.short)
points_64 * points_short
```

###  In-place operations

```python
# In addition to the operations on tensors introduced in the previous section,
# a small number of operations exist only as methods of the Tensor object.
# They are recognizable from a trailing underscore in their name,
# like zero_, which indicates that the method operates in place by modifying the input instead of creating a new output tensor and returning it.
# For instance, the zero_ method zeros out all the elements of the input. 
# Any method without the trailing underscore leaves the source tensor unchanged and instead returns a new tensor:
```

## Tensor metadata: Size, offset, and stride

```python
# The size (or shape, in NumPy parlance) is a tuple indicating the number of elements across each dimension the tensor represents.
# The storage offset is the index in the storage corresponding to the first element in the tensor.
# The stride (ШАГ) is the number of elements in the storage that need to be skipped over to obtain the next element along each dimension.
```

```python
points = torch.tensor([[4.0, 1.0], [5.0, 3.0], [2.0, 1.0]])
second_point = points[1]
```

```python
second_point.storage_offset()
```

```python
second_point.size()
```

```python
second_point.shape
```

```python
# The stride is a tuple indicating the number of elements in the storage that have to be skipped when the index is increased by 1 in each dimension.
```

```python
points.stride()
```

```python
# Accessing an element i, j in a 2D tensor results in accessing the storage_offset + stride[0] * i + stride[1] * j element in the storage.
# The offset will usually be zero; if this tensor is a view of a storage created to hold a larger tensor, the offset might be a positive value.
```

```python
second_point.size()
```

```python
second_point.stride()
```

```python
second_point[0] = 10.0
```

```python
points
```

```python
second_point = points[1].clone() # create dedicated Storage
second_point[0] = 10.1
points
```
