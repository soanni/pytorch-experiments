```python
import torch
import time

device = torch.device("mps")

print("Allocated:", torch.mps.current_allocated_memory() / 1024**2, "MB")
print("Driver allocated:", torch.mps.driver_allocated_memory() / 1024**2, "MB")
print("Recommended memory:", torch.mps.recommended_max_memory() / 1024**3, "GB")


# create large matrices
size = 45000
A = torch.randn(size, size, device=device)
B = torch.randn(size, size, device=device)

print("After tensor creation:")
print("Allocated:", torch.mps.current_allocated_memory()/1024**2, "MB")


# warmup (important for GPU)
for _ in range(3):
    C = A @ B

print("After computation:")
print("Allocated:", torch.mps.current_allocated_memory()/1024**2, "MB")

torch.mps.synchronize()

# benchmark
start = time.time()

C = A @ B

print("After computation:")
print("Allocated:", torch.mps.current_allocated_memory()/1024**2, "MB")

torch.mps.synchronize()
end = time.time()

print("Time:", end - start)
```
