## Broadcasting

```python
# PyTorch will allow us to multiply things that are the same shape, as well as shapes where one operand is of size 1 in a given dimension.
# It also appends the leading dimensions of size 1 automatically.
```

```python
torch.tensor([[1,2,3]]) + 10
```

```python
torch.tensor([[1, 2, 3]]) * torch.tensor([[10],[20],[30]]) # shape(1,3) * shape(3,1) = shape(3,3)
```

```python
torch.tensor([[1, 2, 3], [4, 5, 6]]) + torch.tensor([[[10, 20, 30]], [[40, 50, 60]]]) # shape(2,3) + shape(2, 1, 3) = shape(2,2,3)
```

## Named tensors

```python
img_t = torch.randn(3, 5, 5) # shape(3,5,5) => shape [channels (RGB), rows, columns]
weights = torch.tensor([0.2126, 0.7152, 0.0722])
weights.shape
```

```python
batch_t = torch.randn(2, 3, 5, 5) # shape(2, 3, 5, 5) => shape [batch, channels (RGB), rows, columns]
```

```python
img_gray_naive = img_t.mean(-3) # lazy, unweighted mean on dimension -3 (channel)
img_gray_naive
```

```python
batch_gray_naive = batch_t.mean(-3) # again lazy, unweighted mean on dimension -3 (channel)
batch_gray_naive
```

```python
img_gray_naive.shape, batch_gray_naive.shape
```

```python
unsqueezed_weights = weights.unsqueeze(-1).unsqueeze(-1) # shape(3,1,1)
unsqueezed_weights
```

```python
img_weights = img_t * unsqueezed_weights # shape(3,5,5)
img_weights
```

```python
batch_weights = batch_t * unsqueezed_weights # shape(2,3,5,5) remember that broadcasting appends the leading dimensions of size 1 automatically
batch_weights
```

```python
img_gray_weighted = img_weights.sum(-3) # sum on third dimension from the end (channel), shape(5,5)
img_gray_weighted
```

```python
batch_gray_weighted = batch_weights.sum(-3) # sum on third dimension from the end (channel), shape(2,5,5)
batch_gray_weighted
```

```python
batch_weights.shape, batch_t.shape, unsqueezed_weights.shape
```

```python
weights_named = torch.tensor([0.2126, 0.7152, 0.0722], names = ['channels'])
weights_named
```

```python
print(weights_named.shape, weights_named.names)
```

```python
img_t_named = img_t.refine_names(..., 'channels', 'rows', 'columns')
batch_t_named = batch_t.refine_names(..., 'channels', 'rows', 'columns')
print("img named: ", img_t_named.shape, img_t_named.names)
print("batch named: ", batch_t_named.shape, batch_t_named.names)
```

```python
# For operations with two inputs, in addition to the usual dimension checks—whether sizes are the same, 
# or if one is 1 and can be broadcast to the other—PyTorch will now check the names for us. 
# So far, it does not automatically align dimensions, so we need to do this explicitly.
# The method align_as returns a tensor with missing dimensions added and existing ones permuted to the right order
```

```python
weights_named_aligned = weights_named.align_as(img_t_named)
weights_named_aligned.shape, weights_named_aligned.names
```

```python
# Functions accepting dimension arguments, like sum, also take named dimensions
gray_named = (img_t_named * weights_named_aligned).sum('channels')
gray_named
```

```python
img_t_named[...,:3]
```

```python
# If we want to use tensors outside functions that operate on named tensors, we need to drop the names by renaming them to None.
# The following gets us back into the world of unnamed dimensions
gray_plain = gray_named.rename(None)
gray_plain.shape, gray_plain.names
```
